```python
def full_data(link,category,sub):


    PATH= '/usr/lib/chromium-browser/chromedriver'
    s = Service(PATH)
    driver = webdriver.Chrome(service=s)
    driver.get(link)
    driver.maximize_window()

    time.sleep(14)

    # accept cookies consent
    try:
        consent_button = driver.find_element(By.ID,'onetrust-accept-btn-handler') #ver si el consent button va siempre, creo que no
        consent_button.click()

    except NoSuchElementException:
        pass
    
    time.sleep(7)


    css_selector_button = '.lyracons-search-result-1-x-paginationButtonChangePage.mr3.flex.justify-center'

    
    cambio_pagina = driver.find_elements(By.CSS_SELECTOR,css_selector_button)

    
    links=[link]
    categories=[category]
    subcategories=[sub]
    
    try:
        first_page = driver.find_element(By.CSS_SELECTOR,css_selector_button)
    
        first_page.click()
        url=driver.current_url
     #agrego url
        links.append(url)
        categories.append(category)
        subcategories.append(sub)

    except:
        pass


    while True:
        try:
            time.sleep(7)
            cambio_pagina = driver.find_elements(By.CSS_SELECTOR,css_selector_button)
            cambio_pagina[1].click()
            time.sleep(10)
            links.append(driver.current_url) #agrego urls
            categories.append(category)
            subcategories.append(sub)
            time.sleep(2)
            
            
        except IndexError:
            break

    lista= {'links':links,'categorias':categories,'subcategorias':subcategories}
    df= pd.DataFrame(lista)

    return df
```
